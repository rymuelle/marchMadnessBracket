```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import pprint
import json
```

```python
url = "https://tournament.fantasysports.yahoo.com/t1/pickdistribution"
reponse = requests.get(url)
```

```python
reponse
```

```python
soup = BeautifulSoup(reponse.text, "html.parser")
```

```python
for i, x in enumerate(soup.findAll('script')):
    if "root.App.main" in repr(x): print( i) 
```

```python
sript = soup.findAll('script')[14]
```

```python
data = [line for line in repr(sript).split('\n') if "root.App.main" in line][0]
```

```python
data_json = json.loads(data.replace('root.App.main = ','').replace(';',''))
```

```python
def iterate_through_json(json, depth=0, path_so_far=""):
    for key in json:
        try:
            iterate_through_json(json[key], depth=depth+1, path_so_far=path_so_far+"/"+key)
        except:
            "reached bottom"
            return -1
        buffer =""
        for i in range(depth):
            buffer+=" " 
        print(buffer,key, path_so_far)
        if type(json[key]==str):print(buffer,json[key])
        
```

```python
iterate_through_json(data_json)
```

```python
fantasyGame = data_json["context"]["dispatcher"]["stores"]["TourneyPickDistributionStore"]["state"]["data"]["fantasyGame"]
```

```python
key_to_name_dict = {}
for keys in fantasyGame["tournament"]["tournamentTeams"]:
    print(keys["editorialTeamKey"])
    print(keys["editorialTeam"]['displayName'])
    key_to_name_dict[keys["editorialTeamKey"]] = keys["editorialTeam"]['displayName']
```

```python
key_to_name_dict
```

```python
team_to_percentages = {}

for keys in fantasyGame['pickDistribution']['distributionByRound']:
    #print(keys)
    print(keys['roundId'])
    for x in keys['distributionByTeam']:
        print( x['editorialTeamKey'])
        print( x['percentage'])
        name = key_to_name_dict[x['editorialTeamKey']]
        if name not in team_to_percentages: team_to_percentages[name] = []
        team_to_percentages[name].append(x['percentage'])
        
```
